### Instead of removing the missing values, I replaced the missing values with mean and mode.

```python
# Handle missing values in the updated dataset

# Fill missing numerical values with the mean
numerical_cols_updated = ['card_present_flag', 'merchant_code', 'balance', 'merchant_latitude', 'merchant_longitude']
for col in numerical_cols_updated:
    data[col].fillna(data[col].mean(), inplace=True)

# Fill missing categorical values with the mode
categorical_cols_updated = ['merchant_id', 'merchant_state', 'merchant_suburb', 'txn_description']
for col in categorical_cols_updated:
    data[col].fillna(data[col].mode()[0], inplace=True)

# Verify the missing values have been handled
missing_values_after_filling = data.isnull().sum()

# Display missing values after filling
missing_values_after_filling
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_viz = data

# Step 1: Distribution of Annual Salary
plt.figure(figsize=(10, 6))
sns.histplot(data_viz['annual_salary'], bins=30, kde=True, color='blue')
plt.title('Distribution of Annual Salary')
plt.xlabel('Annual Salary')
plt.ylabel('Frequency')
plt.show()


```

```python
# Step 3: Correlation Heatmap
plt.figure(figsize=(12, 8))
corr_matrix = data_viz[['annual_salary', 'amount', 'balance', 'age']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

```

```python
# Step 5: Box Plot - Annual Salary by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='annual_salary', data=data_viz, palette='Set2')
plt.title('Annual Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Annual Salary')
plt.show()
```

```python
# Adjusting the salary bins based on the observed minimum and maximum values
salary_bins = [29000, 50000, 70000, 100000, 135000]
salary_labels = ['30k-50k', '50k-70k', '70k-100k', '100k-135k']
data['salary_range'] = pd.cut(data['annual_salary'], bins=salary_bins, labels=salary_labels)

# Group the data by salary range and calculate the average spending
salary_spending = data.groupby('salary_range')['amount'].mean()

# Create a bar chart to visualize average spending based on annual salary ranges
plt.figure(figsize=(12, 6))
plt.bar(salary_spending.index, salary_spending.values, color='orange')
plt.xlabel('Annual Salary Range')
plt.ylabel('Average Spending Amount')
plt.title('Average Spending Patterns Based on Annual Salary Range')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Grouping the dataset by city and calculating the average spending amount
# The dataset doesn't have a direct 'city' column, so using latitude and longitude to approximate cities if available

# Checking unique values in latitude and longitude to determine if cities can be differentiated
location_summary = data[['latitude', 'longitude']].drop_duplicates()
location_summary.head()
# Mapping latitude and longitude to city names based on known coordinates
def map_city(latitude, longitude):
    if -28 <= latitude <= -27 and 153 <= longitude <= 154:
        return 'Brisbane'
    elif -34 <= latitude <= -33 and 151 <= longitude <= 152:
        return 'Sydney'
    elif -38 <= latitude <= -37 and 144 <= longitude <= 145:
        return 'Melbourne'
    elif -35 <= latitude <= -34 and 138 <= longitude <= 139:
        return 'Adelaide'
    elif -32 <= latitude <= -31 and 115 <= longitude <= 116:
        return 'Perth'
    else:
        return 'Other'

# Apply the city mapping function to create a new column 'city'
data['city'] = data.apply(lambda row: map_city(row['latitude'], row['longitude']), axis=1)

# Group the data by city and calculate the average spending
city_spending = data.groupby('city')['amount'].mean()

# Create a bar chart to visualize average spending based on city
plt.figure(figsize=(12, 6))
plt.bar(city_spending.index, city_spending.values, color='teal')
plt.xlabel('City')
plt.ylabel('Average Spending Amount')
plt.title('Average Spending Patterns by City')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group the data by hour and calculate the average spending amount
hourly_spending = data.groupby('hour')['amount'].mean()

# Create a line chart to visualize the average spending amount for each hour of the day
plt.figure(figsize=(10, 6))
plt.plot(hourly_spending.index, hourly_spending.values, marker='o', linestyle='-', linewidth=2, color='purple')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Spending Amount')
plt.title('Average Spending Patterns by Hour')
plt.grid(True)
plt.xticks(range(0, 24))
plt.show()
```

```python
# Checking the unique values in the 'txn_description' column to understand the different payment methods
payment_methods = data['txn_description'].unique()
payment_methods
# Group the data by payment method and calculate the average spending
payment_spending = data.groupby('txn_description')['amount'].mean()

# Create a bar chart to visualize average spending based on payment method
plt.figure(figsize=(12, 6))
plt.bar(payment_spending.index, payment_spending.values, color='coral')
plt.xlabel('Payment Method')
plt.ylabel('Average Spending Amount')
plt.title('Average Spending Patterns by Payment Method')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Define age bins to categorize the age groups
age_bins = [18, 25, 35, 45, 55, 65, 75, 85]
age_labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '66-75', '76-85']
data['age_group'] = pd.cut(data['age'], bins=age_bins, labels=age_labels)

# Group the data by age group and count the number of transactions
age_group_distribution = data['age_group'].value_counts().sort_index()

# Create a bar chart to visualize the distribution of transactions by age group
plt.figure(figsize=(10, 6))
plt.bar(age_group_distribution.index, age_group_distribution.values, color='blue')
plt.xlabel('Age Group')
plt.ylabel('Number of Transactions')
plt.title('Distribution of Transactions by Age Group')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Grouping the data by city and calculating the total spending to find the highest spending cities
city_total_spending = data.groupby('city')['amount'].sum().sort_values(ascending=False)


# Visualizing the highest spending cities
plt.figure(figsize=(12, 6))
plt.bar(city_total_spending.index, city_total_spending.values, color='green')
plt.xlabel('City')
plt.ylabel('Total Spending Amount')
plt.title('Total Spending by City')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
# Grouping the data by city and counting the number of transactions for each city
city_transaction_volume = data['city'].value_counts()

# Creating a bar chart to visualize the transaction volumes across different cities
plt.figure(figsize=(12, 6))
plt.bar(city_transaction_volume.index, city_transaction_volume.values, color='purple')
plt.xlabel('City')
plt.ylabel('Number of Transactions')
plt.title('Transaction Volumes Across Different Cities')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Grouping the data by transaction type ('txn_description') and calculating the total spending for each type
transaction_type_spending = data.groupby('txn_description')['amount'].sum().sort_values(ascending=False)

# Creating a bar chart to visualize the highest spending by transaction type
plt.figure(figsize=(12, 6))
plt.bar(transaction_type_spending.index, transaction_type_spending.values, color='orange')
plt.xlabel('Transaction Type')
plt.ylabel('Total Spending Amount')
plt.title('Total Spending by Transaction Type')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Grouping the data by gender and age to calculate total spending for each combination
gender_age_spending = data.groupby(['gender', 'age'])['amount'].sum().reset_index()

# Creating a pivot table for better visualization
gender_age_pivot = gender_age_spending.pivot(index='age', columns='gender', values='amount')

# Creating a line chart for each gender to visualize peak spending by age
plt.figure(figsize=(12, 6))
for gender in gender_age_pivot.columns:
    plt.plot(gender_age_pivot.index, gender_age_pivot[gender], marker='o', linestyle='-', linewidth=2, label=f'Gender: {gender}')

plt.xlabel('Age')
plt.ylabel('Total Spending Amount')
plt.title('Total Spending by Gender and Age')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Grouping the data by salary range to calculate total spending for each salary category
salary_spending_total = data.groupby('salary_range')['amount'].sum().sort_index()

# Creating a bar chart to visualize total spending based on salary ranges
plt.figure(figsize=(12, 6))
plt.bar(salary_spending_total.index, salary_spending_total.values, color='blue')
plt.xlabel('Salary Range')
plt.ylabel('Total Spending Amount')
plt.title('Total Spending by Salary Range')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Grouping the data by day of the week and calculating the total spending for each day
day_spending_total = data.groupby('dow')['amount'].sum().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Creating a bar chart to visualize total spending by day of the week
plt.figure(figsize=(10, 6))
plt.bar(day_spending_total.index, day_spending_total.values, color='magenta')
plt.xlabel('Day of the Week')
plt.ylabel('Total Spending Amount')
plt.title('Total Spending by Day of the Week')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Filtering the data for weekends (Saturday and Sunday)
weekend_data = data[data['dow'].isin(['Saturday', 'Sunday'])]

# Calculating the average spending on weekends
average_weekend_spending = weekend_data['amount'].mean()

# Filtering the data for weekdays (Monday to Friday)
weekday_data = data[~data['dow'].isin(['Saturday', 'Sunday'])]

# Calculating the average spending on weekdays
average_weekday_spending = weekday_data['amount'].mean()

# Creating a bar chart to compare average spending on weekends and weekdays
plt.figure(figsize=(8, 6))
plt.bar(['Weekdays', 'Weekends'], [average_weekday_spending, average_weekend_spending], color=['blue', 'orange'])
plt.xlabel('Day Type')
plt.ylabel('Average Spending Amount')
plt.title('Comparison of Average Spending on Weekdays vs. Weekends')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Converting the 'date' column to datetime format for better time-based analysis
data['date'] = pd.to_datetime(data['date'], errors='coerce')

# Grouping the data by date and counting the number of transactions per day
daily_transaction_volume = data.groupby('date').size()

# Creating a line chart to visualize transaction volumes over time
plt.figure(figsize=(12, 6))
plt.plot(daily_transaction_volume.index, daily_transaction_volume.values, color='darkcyan', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.title('Transaction Volumes Over Time')
plt.grid(True)
plt.show()
```

```python
# Grouping the data by city to calculate average spending for each location
location_spending = data.groupby('city')['amount'].mean().sort_values(ascending=False)

# Creating a bar chart to visualize average spending based on location (city)
plt.figure(figsize=(12, 6))
plt.bar(location_spending.index, location_spending.values, color='brown')
plt.xlabel('City')
plt.ylabel('Average Spending Amount')
plt.title('Average Spending Patterns Based on Location (City)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
